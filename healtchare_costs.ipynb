```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
df = pd.read_excel('./healthcare.costs.xlsx', sheet_name='Health', index_col='Country')

# Change col names for ease of reference
df.columns = ['region', '2020pcc', 'obesity_pct']
df.index.name = 'country'
```

```python
# Find US for chart annotation
df[df.index == 'United States']
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.regplot(data=df, x='2020pcc', y='obesity_pct', marker=None, line_kws={'color': 'black'})
sns.scatterplot(data=df, x='2020pcc', y='obesity_pct', hue='region', palette='deep', s=80)
plt.annotate('United States', xy=(11859, .3647), xytext=(10000, .35), arrowprops=dict(color='black', arrowstyle='->'))

ax.xaxis.set_major_formatter('${x:,.0f}')
ax.yaxis.set_major_formatter('{x:.0%}')
plt.legend(loc='upper left', facecolor='white')
plt.title('Obesity Rate Percent vs. Healthcare Costs Per Capita')
plt.xlabel(None, fontsize=12)
plt.ylabel(None, fontsize=12)
plt.show();
```
